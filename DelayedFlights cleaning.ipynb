```python

df = pd.read_csv('DelayedFlights.csv')
print(df.shape)
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
backup_df = df.copy()
```

```python
df['DepTime'] = df['DepTime'].replace(0, pd.NA)
df['ArrTime'] = df['ArrTime'].replace(0, pd.NA)
df['CRSDepTime'] = df['CRSDepTime'].replace(0, pd.NA)
df['CRSArrTime'] = df['CRSArrTime'].replace(0, pd.NA)
time_cols = ['DepTime', 'ArrTime', 'CRSDepTime', 'CRSArrTime']

for col in time_cols:
    df[col] = df[col].apply(lambda x: f"{int(x):04}" if pd.notna(x) else x)
    df[col] = pd.to_datetime(df[col], format='%H%M', errors='coerce').dt.time
print(df.head())
```

```python
df.dropna(subset=['TailNum'], inplace=True)
df['TailNum'].isnull().sum()
```
